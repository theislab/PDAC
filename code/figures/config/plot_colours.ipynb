```python
import yaml
import matplotlib.pyplot as plt

with open("config.yml", "r") as f:
    config = yaml.safe_load(f)


```

```python

for palette_name, palette in config.get("palettes", {}).items():
    if not isinstance(palette, dict):
        continue  
    if palette_name in ["Level_4_Daniele_2","Level_4_old"]:
        continue

    fig, ax = plt.subplots(figsize=(8, len(palette) * 0.4))

    for i, (label, color) in enumerate(palette.items()):
        ax.add_patch(plt.Rectangle((0, i), 1, 1, color=color))
        ax.text(1.1, i + 0.5, label, va='center', ha='left', fontsize=9)

    ax.set_xlim(0, 2)
    ax.set_ylim(0, len(palette))
    ax.set_title(palette_name, fontsize=14, loc='left')
    ax.axis('off')
    plt.tight_layout()
    plt.show()
```
